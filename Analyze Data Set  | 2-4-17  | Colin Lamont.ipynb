## Housing Data Set and Related Statistics of Boston Suburbs, 1978
#### Sourced from: https://archive.ics.uci.edu/ml/datasets/Housing
#### Originally published by Harrison, D. and Rubinfeld, D.L. in 'Hedonic prices and the demand for clean air', J. Environ. Economics & Management, vol.5, 81-102, 1978.
#### Contains 506 instances across 14 attributes:

- CRIM: per capita crime rate by town 
- ZN: proportion of residential land zoned for lots over 25,000 sq.ft. 
- INDUS: proportion of non-retail business acres per town 
- CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise) 
- NOX: nitric oxides concentration (parts per 10 million) 
- RM: average number of rooms per dwelling 
- AGE: proportion of owner-occupied units built prior to 1940 
- DIS: weighted distances to five Boston employment centres 
- RAD: index of accessibility to radial highways 
- TAX: full-value property-tax rate per 10,00 Dollars
- PTRATIO: pupil-teacher ratio by town 
- B: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town 
- LSTAT: % lower status of the population 
- MEDV: Median value of owner-occupied homes in 1000's of Dollars

```python
import numpy as py
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
bd = pd.read_csv("boston_housing_data.csv")
```

```python
bd
```

### Finding the Median for Home Values in Boston Suburbs, in Dollars

```python
medv_med = bd['MEDV'].median()
```

```python
print("Median: ${0:,.2f}".format(medv_med * 1000))
```

### Finding the Mean for Home Values in Boston Suburbs, in Dollars

```python
medv_mean = bd['MEDV'].mean()
```

```python
print("Mean: ${0:,.2f}".format(medv_mean * 1000))
```

### Finding the Median Home Value in the Boston Suburb with the youngest houses

```python
bd_age = bd['AGE'].min()
```

```python
the_row = bd.loc[bd.AGE == bd_age].MEDV.values[0]
```

```python
print("Median home value on the section of Boston homes with the youngest age: ${0:,.2f}".format(the_row * 1000))
```

### Correlation between Crime Rates and Student-Teacher Ratios

```python
x = bd.CRIM
y = bd.PTRATIO
```

```python
plt.plot(bd[bd.CHAS == 0].CRIM, bd[bd.CHAS == 0].PTRATIO, 'go')
plt.plot(bd[bd.CHAS == 1].CRIM, bd[bd.CHAS == 1].PTRATIO, 'ro')
plt.xlabel("Crime Rate Per-Capita by Boston Suburb")
plt.ylabel("Student-Teacher Ratio by Boston Suburb")
plt.title("Correlation between per-capita crime rate and the student-teacher ratio. \nRed denotes suburbs bounded by the Charles River.")
plt.show()
```

### Correlation between Black Population and Distance from Boston Business Centers

```python
plt.plot(bd.B, bd.DIS, 'bo')
plt.xlabel("Black Polulation Ratio in Boston Suburb")
plt.ylabel("Distance from Boston Business Centers")
plt.title("Correlation between Black Population and Proximity to Boston Business Centers.")
plt.show()
```

### Median Home Values correlated with Nitric Oxide levels and Industrial Business Concentration

```python
plt.plot(bd.MEDV, bd.NOX, 'bo', bd.MEDV, bd.INDUS, 'ro')
plt.xlabel("Median Home Values per Boston Suburb (blue)")
plt.ylabel("Nitric oxide concentration (blue)\nNon-retail businesses per suburb (red)")
plt.title("Correlation between Median Home Values against both Nitric Oxide Levels and Non-Tetail Businesses")
plt.show()
```

### The Correlation between Nitric Oxide levels and Industrial Areas
