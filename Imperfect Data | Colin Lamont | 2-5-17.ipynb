## Imperfect Data: Cleaning and Analyzing NYC SAT Scores
### 2012 College-Bound Graduating HS Seniors from NYC
#### Data pulled from: https://data.cityofnewyork.us/Education/SAT-Results/f9bf-2cp4

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sat = pd.read_csv("../SAT_Results.csv")
```

```python
sat
```

```python
sat.columns = ['DBD', 'NAME', 'Num', 'Read', 'Math', 'Write']
```

```python
sat.columns
```

```python
sat
```

## #Remove the null rows, where the number and score fields are represented with an "s"

```python
cs = sat[sat.Read != "s"]
```

```python
cs
```

### Plot the expected high correlation between reading and writing SAT scores

```python
plt.plot(cs.Read, cs.Write, 'go')
plt.xlabel("Reading SAT Scores")
plt.ylabel("Writing SAT Scores")
plt.title("The extremely high correlation between reading and writing SAT scores")
plt.show()
```

### Get the Reading Max, Min, and Median Scores

```python
read_max = cs.Read.max()
read_min = cs.Read.min()
read_med = cs.Read.median()
```

```python
print("The highest reading score was: {0:}".format(read_max))
print("The lowest reading score was: {0:}".format(read_min))
print("The median reading score was: {0:}".format(read_med))
```

### Get the Writing Max, Min, and Median Scores

```python
write_max = cs.Write.max()
write_min = cs.Write.min()
write_med = cs.Read.median()
```

```python
print("The highest writing score was: {0:}".format(write_max))
print("The lowest writing score was: {0:}".format(write_min))
print("The median writing score was: {0:}".format(write_med))
```

### Get the Math Max, Min, and Median Scores

```python
math_max = cs.Math.max()
math_min = cs.Math.min()
math_med = cs.Math.median()
```

```python
print("The highest match score was: {0:}".format(math_max))
print("The lowest math score was: {0:}".format(math_min))
print("The median math score was: {0:}".format(math_med))
```

### Get the Average scores for Math, Reading, and Writing SATs

```python
cs.Math = cs.Math.astype("int")
```

```python
math_avg = cs.loc[cs.Math < 800].Math.values.mean()
```

```python
print("The average math SAT score was: {0:.1f}".format(math_avg))
```

```python
cs.Read = cs.Read.astype("int")
```

```python
read_avg = cs.loc[cs.Read < 800].Read.values.mean()
```

```python
print("The average reading SAT score was: {0:.1f}".format(read_avg))
```

```python
cs.Write = cs.Write.astype("int")
```

```python
write_avg = cs.loc[cs.Write < 800].Write.values.mean()
```

```python
print("The average writing SAT score was: {0:.1f}".format(write_avg))
```

### Get the total number of students who took the SAT in 2012 in NYC

```python
cs.Num = cs.Num.astype("int")
```

```python
students = cs.loc[cs.Num > 0].Num.values.sum()
```

```python
print("Number of students who took the SAT in NYC in 2012: {0:,}".format(students))
```
